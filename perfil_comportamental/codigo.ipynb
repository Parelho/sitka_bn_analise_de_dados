```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dados = pd.read_csv(os.path.join("dados.csv"))
display(dados)
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages  # Import PdfPages

import os

custom_font = fm.FontProperties(fname=os.path.join("Now-Regular.ttf"))

# Create a PDF file to save all the pie charts
pdf_filename = 'graficos_perfil_comportamental.pdf'
pdf_pages = PdfPages(pdf_filename)

# Assuming 'dados' is your DataFrame
for index, row in dados.iterrows():
    valores = row[['Inovador', 'Comunicador', 'Executor', 'Planejador']].values.flatten()
    Inovador = row["Inovador"]
    Comunicador = row["Comunicador"]
    Planejador = row["Executor"]
    Executor = row["Planejador"]
    
    plt.figure(figsize=(8, 6))
    
    wedges, autotexts = plt.pie(
        valores, startangle=90, labeldistance=1.5,
        labels=(f"{round(Inovador*100)}% | Inovador    \n(Melhoria Continua)", f"{round(Comunicador*100)}% | Comunicador\n(Relacionamento) ", f"{round(Planejador*100)}% | Planejador\n(Ordem/Controle)", f"{round(Executor*100)}% | Executor\n(Atitude/Ação)"), 
        colors=["#EBECED", "#BDBFBF", "#E21726","#173D70"], 
        textprops={'fontproperties': custom_font, 'color': '#173D70', 'fontsize': 35})
    
    for autotext in autotexts:
        autotext.set_fontproperties(custom_font)
        autotext.set_fontsize(20)

    nome = row["Nome"]
    plt.suptitle(f'{nome}', y=1.05, fontproperties=custom_font, fontsize=16)
    plt.tight_layout()

    pdf_pages.savefig(bbox_inches='tight')

    plt.close()

pdf_pages.close()
```
